## Creating Arrays

```python
# create an array of all zeros
np.zeros(5) # creates an array of 5 0's
```

```python
np.ones(10) # creates an array of 10 1's
```

```python
np.full(10,2.5) # creates an array of 10 2.5's
```

## Converting Lists to Arrays

```python
my_list = [2,3,4]
array_from_list = np.array(my_list)
array_from_list
```

```python
# get the second item from the array
array_from_list[1] # remember python 0 indexes!
```

```python
# reassign the second item of the array to a new value
array_from_list[1] = 10
array_from_list
```

## Generating Ranges of Numbers

```python
range_array = np.arange(10) # create an array from 0 to 9
range_array
```

```python
# you can also se the specified start number
short_range_array = np.arange(3,10) # creates an array from 3 (inclusive) to 10 (exclusive)
short_range_array
```

```python
# create an array evenly spaced numbers within a range
linspace_array = np.linspace(0,1,11) # creates 11 numbers from 0 to 1
linspace_array
```

## Multi-Dimensional Arrays

```python
zeroes_matrix = np.zeros((5,2))
zeroes_matrix # an array with 5 rows and 2 columns filled with zeroes (shape is (5,2)
```

```python
ones_matrix = np.ones((5,2))
ones_matrix # (5,2) array filled with ones
```

## Array Index and Slicing

```python
arr = np.array([[2,3,4],
                [4,5,6]])
first_row = arr[0]
first_col = arr[:, 0]

print(first_row)
print(first_col)
```

```python
# get the 3rd number of the second row
arr[1,2]
```

## Randomly Generated Arrays

```python
np.random.seed(2) # specifiy the seed for reproducibility
random_array = np.random.rand(5,2)
random_array
```

```python
# random numbers from a normal distribution
np.random.seed(2)
normal_dist = np.random.randn(5,2)
normal_dist
```

```python
# random integers within a range
np.random.seed(2)
rand_ints = np.random.randint(low= 0, high=100, size=(5,2))
rand_ints
```

## Array Operations

### Element-Wise Operations

```python
# create an array for demo
a = np.arange(5)
a
```

```python
# add 1 to each member of the array
a = a + 1
a
```

```python
# multiply each element by 2
a = a * 2
a
# works the same for other mathematical operations like division etc
```

### Element-Wise Operations with Two Arrays

```python
arr1 = np.ones(4)
arr1
```

```python
arr2 = np.full(4,3) # creates an array of length 4 full of 3's
arr2
```

```python
result1 = arr1 + arr2 # adds each element together element wise
result1
```

```python
result2 = arr1 / arr2 # divides each number
result2
```

### Comparison Operations

```python
# perform element-wise comparison
arr = np.array([1,2,3,4])
greater_than_2 = arr > 2 # produces an array of booleans representing rather that element is greater than 2 or not
greater_than_2
```

### Selecting Elements Based on Conditions

```python
selection = arr[arr>2] # selects elements that are True
selection
```

## Summary Operations

```python
# minimum value
arr.min()
```

```python
# max val
arr.max()
```

```python
# sum of all elements
arr.sum()
```

```python
# mean of all elements
arr.mean()
```

```python
# standard deviation of all elements
arr.std()
```
